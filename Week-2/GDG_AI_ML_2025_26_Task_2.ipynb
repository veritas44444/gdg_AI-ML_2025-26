```python
import pandas as pd
house_table= pd.read_csv("Housing.csv")
house_table.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
#accessing the relevant dataset



house_table.info()
house_table.describe()
Area = house_table["area"]
Price = house_table["price"]
house_table[["area", "price"]]
#stating portions for training and testing model
Area_train, Area_test, Price_train, Price_test = train_test_split(Area, Price, test_size=0.2, random_state=0)

H_Price_Model = LinearRegression()
# Reshape Area_train for single feature input as LinearRegression expects a 2D array
H_Price_Model.fit(Area_train.values.reshape(-1, 1), Price_train)

predictions = H_Price_Model.predict(Area_test.values.reshape(-1, 1))

print("Mean Squared Error:", mean_squared_error(Price_test, predictions))
print("Mean Absolute Error:", mean_absolute_error(Price_test, predictions))
```
